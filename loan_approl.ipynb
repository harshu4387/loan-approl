```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Reload the correct dataset to ensure 'loan_amount' column is available
df = pd.read_csv('/content/loan_approval_dataset.csv')

# Clean column names by stripping any leading/trailing whitespace
df.columns = df.columns.str.strip()

# Create a box plot for 'loan_amount'
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['loan_amount'])

# Set the title and labels
plt.title('Distribution of Loan Amount')
plt.ylabel('Loan Amount')

# Display the plot
plt.show()

print("Box plot for Loan Amount displayed successfully.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define the list of available categorical features to analyze
categorical_features = ['education', 'self_employed'] # Adjusted based on available columns

for feature in categorical_features:
    # Calculate the proportion of approved loans for each category
    approval_rates = df.groupby(feature)['loan_status'].value_counts(normalize=True).unstack()

    # Ensure 'Approved' column exists before proceeding
    if 'Approved' in approval_rates.columns:
        approval_rates['Approval_Rate'] = approval_rates['Approved'] * 100

        # Create a bar plot
        plt.figure(figsize=(8, 6))
        sns.barplot(x=approval_rates.index, y=approval_rates['Approval_Rate'], palette='viridis')

        # Set the title and labels
        plt.title(f'Loan Approval Rate by {feature.replace("_", " ").title()}')
        plt.xlabel(feature.replace("_", " ").title())
        plt.ylabel('Loan Approval Rate (%)')

        # Display the plot
        plt.show()
    else:
        print(f"'Approved' status not found in '{feature}' group. Skipping plot for this feature.")

print("Bar charts for loan approval rates by 'Education' and 'self_employed' displayed successfully.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure 'loan_status' column values are clean by stripping whitespace
df['loan_status'] = df['loan_status'].str.strip()

# Define the list of available categorical features to analyze
categorical_features = ['education', 'self_employed']

for feature in categorical_features:
    # Calculate the proportion of approved loans for each category
    approval_rates = df.groupby(feature)['loan_status'].value_counts(normalize=True).unstack()

    # Ensure 'Approved' column exists before proceeding
    if 'Approved' in approval_rates.columns:
        approval_rates['Approval_Rate'] = approval_rates['Approved'] * 100

        # Create a bar plot
        plt.figure(figsize=(8, 6))
        sns.barplot(x=approval_rates.index, y=approval_rates['Approval_Rate'], palette='viridis')

        # Set the title and labels
        plt.title(f'Loan Approval Rate by {feature.replace("_", " ").title()}')
        plt.xlabel(feature.replace("_", " ").title())
        plt.ylabel('Loan Approval Rate (%)')

        # Display the plot
        plt.show()
    else:
        print(f"'Approved' status not found in '{feature}' group even after stripping. Skipping plot for this feature.")

print("Bar charts for loan approval rates by 'Education' and 'self_employed' displayed successfully.")
```
